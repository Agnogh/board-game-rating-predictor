## Check Missing Values

In this section, I'll check which columns contain missing values.

Missing values are important because they can affect mz analysis and machine learning models. Before cleaning the data, we first need to first understand where the missing values are and how many there is.

```python
# Count missing values in each column.
df.isnull().sum()
```

```python
# Calculate missing values as percenteg
missing_percentege = (df.isnull().sum() / len(df)) * 100

missing_percentege
```

```python
# Create a clerer missing values summary table
missing_summary = pd.DataFrame({
    "missing_count": df.isnull().sum(),
    "missing_percentege": (df.isnull().sum() / len(df)) * 100
})

# Show only columns that have missing values.
missing_summary[missing_summary["missing_count"] > 0]
```

### Missing Values Observations

The missing values check shows that 5 columns contain missing data.

Columns with missing values are:

- `ID`: 16 missing values, approximetely 0.08%
- `Year Published`: 1 missing value, approximetely 0.005%
- `Owned Users`: 23 missing values, approximetely 0.11%
- `Mechanics`: 1,598 missing values, approximetely 7.86%
- `Domains`: 10,159 missing values, approximetely 49.94%

The `ID`, `Year Published`, and `Owned Users` columns have very low levels of missing data.

The `Mechanics` column has a moderate amount of missing data.

The `Domains` column has a very high amount of missing data, with almost half of the values missing. This column may need special handling during the data cleaning and feature selection stages.

At this stage, the missing values are only being identified, not fixed.

## Check Duplicate Records

In this section, we check whether the dataset contains duplicate records.

Duplicate records matters because repeated rows can distort analysis and machine learning results.

```python
# Count fully duplicated rows (check repetations)
duplicate_rows = df.duplicated().sum()

duplicate_rows
```

```python
# Check whether any board game names appear more than once (versions/editions of same game)
duplicate_names = df["Name"].duplicated().sum()

duplicate_names
```

```python
# Show examples of duplicated game names, if case any exists
df[df["Name"].duplicated(keep=False)].sort_values("Name").head(20)
```

### Duplicate Records Observations

The dataset contains 0 fully 100% duplicated rows.

There are 367 repeated occurrences in the `Name` column. However, repeated game names does not necessarily represent duplicate records. The examples show that games with the same name may have different IDs, publication (release) years, player counts, ratings, or mechanics.

These records may represent different editions, versions, or separate games that share the same title. Therefore, duplicated names should not be removed automatically and will need further review during data cleaning.

## Prediction Target and Candidate Features

The main prediction target for this project is:

* `Rating Average`

This means that the machine learning model will attempt to predict the average user rating of a board game.

Initial candidate features include:

* `Year Published`
* `Min Players`
* `Max Players`
* `Play Time`
* `Min Age`
* `Complexity Average`
* `Mechanics`
* `Domains`

Some columns will not initially be used as prediction features:

* `ID` is an identifier and does not describe the characteristics of a game.
* `Name` identifies the game but is not initially sutable as a direct numeric feature.
* `BGG Rank` is strongly connected to game ratings and could cause data leakage.
* `Users Rated` and `Owned Users` describe popularity after release and may not be available when predicting the rating of a new game.

The final feature list may change after further exploration, cleaning, and model testing.


```python
# To define the value that the model will try to predict
target_column = "Rating Average"

# To define initial columns that may be used as model features
candidate_features = [
    "Year Published",
    "Min Players",
    "Max Players",
    "Play Time",
    "Min Age",
    "Complexity Average",
    "Mechanics",
    "Domains"
]

print("Prediction target:")
print(target_column)

print("\nCandidate features:")
for feature in candidate_features:
    print("-", feature)
```

```python
# Select only numeric columns, then calculate their 
# correlation with prediction target
numeric_correlations = (
    df.select_dtypes(include="number")
    .corr()["Rating Average"]
    .sort_values(ascending=False)
)

numeric_correlations
```

### Prediction Target and Correlation Observations

The main prediction target is `Rating Average`.

The initial candidate features describe characteristics of the board game, including publication (release) year, player counts, play time, minimum age, complexity, mechanics, and domains.

Among the numeric candidate features, `Complexity Average` has the strongest positive correlation with `Rating Average`, at approximately 0.48. This suggests a moderate relationship between game complexity and average user rating.

`Min Age`, `Play Time`, and `Year Published` have weaker positive correlations with the target. `Min Players` has a small negative correlation, while `Max Players` has very little linear relationship with the target.

`BGG Rank` has a strong negative correlation with `Rating Average`. However, it will not be used as a prediction feature because game rank is already strongly influenced by ratings.

`ID` also shows a positive correlation, but it is only an identifier and does not represent a any meaningful board game characteristic.

`Users Rated` and `Owned Users` describe popularity after release and will not initially be used when predicting ratings from a game's characteristics.

The text-based `Mechanics` and `Domains` columns do not appear in the numeric correlation results. They will require transformation before they can be used by a machine learning model.

Correlation measures only simple linear relationships and does not determine the final usefulness of a feature.
